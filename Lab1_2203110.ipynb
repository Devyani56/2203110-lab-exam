```python
import numpy as np
n=10000
p1=0.2

def simulate_dataset(n,p1):
  vector=[]
  for _ in range(n):
    value=1 if np.random.random()<p1 else 0
    vector.append(value)
  return vector

dataset = simulate_dataset(n,p1)
print(dataset)
```

```python
p=p1
def random_classification(n,p):
  vector=[]
  for _ in range(n):
    value=1 if np.random.random()<p else 0
    vector.append(value)
  return vector

randomly_predicted = random_classification(n,p)
print(randomly_predicted)
```

```python
def bayes_classifier(n,p1):

  vector=[1]*n if p1>=0.5 else [0]*n
  return vector

bayes_predicted = bayes_classifier(n,p1)
print(bayes_predicted)
```

```python
def expected_zero_one_loss(n,p1,classifier):
  dataset = simulate_dataset(n,p1)
  predicted = classifier(n,p1)
  differences = sum(1 for a, b in zip(dataset, predicted) if a != b)
  return differences/n

print(expected_zero_one_loss(n,p1,random_classification))
print(expected_zero_one_loss(n,p1,bayes_classifier))
```

```python
def true_zero_one_loss(n,p1,classifier,p):
     if classifier == bayes_classifier:
      return min(p,1-p)
     else:
      return p*(1-p1)+(1-p)*p1
print(true_zero_one_loss(n,p1,random_classification,p))
print(true_zero_one_loss(n,p1,bayes_classifier,p))
```

```python
n=10000
print("Bayes Classifier:")
print("Expected zero-one loss:", expected_zero_one_loss(n,p1,bayes_classifier),"True zero-one loss:", true_zero_one_loss(n,p1,bayes_classifier,p))
print("Random Classifier:")
print("Expected zero-one loss:", expected_zero_one_loss(n,p1,random_classification),"True zero-one loss:", true_zero_one_loss(n,p1,random_classification,p))
```

```python
p=0.1
x0_y1=0.1
x1_y1=0.9
x0_y0 =0.95
x1_y0=0.05
def generate_feature_vector(input_vector,x0_y1,x0_y0):
    new_vector = []
    for value in input_vector:


        if value == 1:
            rand_num = np.random.random()  # Generate random number between 0 and 1
            if rand_num <= x0_y1:
                new_vector.append(0)
            else:
                new_vector.append(1)
        elif value == 0:
            rand_num = np.random.random()  # Generate random number between 0 and 1
            if rand_num <= x0_y0:
                new_vector.append(0)
            else:
                new_vector.append(1)

    return new_vector

def new_bayes_classifier(n,p,dataset):
  feature_vector = generate_feature_vector(dataset,x0_y1,x0_y0)
  result_vector=[]
  for value in feature_vector:
    if value == 1:
      if x1_y0*(1-p)> x1_y1*p:
        result_vector.append(0)
      else:
        result_vector.append(1)
    else:
      if x0_y0*(1-p)> x0_y1*p:
        result_vector.append(0)
      else:
        result_vector.append(1)

  return result_vector


dataset = simulate_dataset(n,p)
#positive= dataset.count(1)
#negative= dataset.count(0)
#py_zero = negative/n
#py_one = positive/n
predicted_bayes = new_bayes_classifier(n,p1,dataset)

def type_one_error(dataset,predicted,n):
 count = 0
 for a, b in zip(dataset, predicted):
     if a == 0 and b == 1:
            count += 1
 return count/n
def type_two_error(dataset,predicted,n):
 count = 0
 for a, b in zip(dataset, predicted):
        if a == 1 and b == 0:
            count += 1
 return count/n
print("Bayes Classifier - Type1 loss=",type_one_error(dataset,predicted_bayes,n),"Type2 loss=",type_two_error(dataset,predicted_bayes,n))
```
